```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_LC():
    
    #Plotting information
    fig, axs = plt.subplots(ncols = 4, nrows = 2, figsize = (7, 3), dpi = 250)
    position = [axs[0, 0], axs[0, 1], axs[0, 2], axs[0, 3],
                axs[1, 0], axs[1, 1], axs[1, 2], axs[1, 3]]


    
    #Plotting
    for i in range(4):
        ax = position[i]
        ax.plot(globals()['pso_' + str(i)]['Samples'], globals()['pso_' + str(i)]['MAE'], marker = 'o', color = 'black', markersize = 2, linewidth = 0.5, label = 'PSO')
        ax.plot(globals()['bo_' + str(i)]['Samples'], globals()['bo_' + str(i)]['MAE'], marker = 'o', color = 'red', markersize = 2, linewidth = 0.5, label = 'BBO')
        ax.plot(globals()['grid_' + str(i)]['Samples'], globals()['grid_' + str(i)]['MAE'], marker = 'o', color = 'blue', markersize = 2, linewidth = 0.5, label = 'Grid')
        ax.set_ylim(-0.01, 0.205)
        ax.set_yticks([0, 0.1, 0.2], fontsize = 6)
        ax.set_ylabel('MAE')
        if i > 0:
            ax.set_yticks([])
            ax.set_ylabel('')
        
        if i == 3:
            ax.set_xlim(16, 165)
            ax.set_xticks([24, 160], fontsize = 6)
        else:
            ax.set_xlim(10, 105)
            ax.set_xticks([15, 100], fontsize = 6)
            
        if i < 4:
            ax.set_xticks([])
        
        ax.set_title(f'Model {i+1}')
        
        #2nd row plot
        ax1 = position[4 + i]
        ax1.plot(globals()['pso_' + str(i)]['Samples'], globals()['pso_' + str(i)]['R_square'], marker = 'o', color = 'black', markersize = 2, linewidth = 0.5, label = 'PSO')
        ax1.plot(globals()['bo_' + str(i)]['Samples'], globals()['bo_' + str(i)]['R_square'], marker = 'o', color = 'red', markersize = 2, linewidth = 0.5, label = 'BBO')
        ax1.plot(globals()['grid_' + str(i)]['Samples'], globals()['grid_' + str(i)]['R_square'], marker = 'o', color = 'blue', markersize = 2, linewidth = 0.5, label = 'Grid')
        ax1.set_ylim(0., 1.05)
        ax1.set_yticks([0., 0.5, 1], fontsize = 6)
        ax1.set_ylabel('$R^2$')
        if i > 0:
            ax1.set_yticks([])   
            ax1.set_ylabel('')
        if i == 3:
            ax1.set_xlim(16, 165)
            ax1.set_xticks([24, 160], fontsize = 6)
        else:
            ax1.set_xlim(10, 105)
            ax1.set_xticks([15, 100], fontsize = 6)
            
            

    plt.legend(ncol = 3, bbox_to_anchor = (-0.2, -0.35))

    #Labelling
    plt.text(-160, -0.3, s='Number of Samples', fontsize = 10, ha='center', va='center')
    plt.subplots_adjust(wspace=0.12, hspace=0.05)

    plt.show()
    return
```
